```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("IMDB Dataset.csv", on_bad_lines='skip')
```

```python
#lowercase: str.lower() pandas string fonksiyonu
df['review'] = df['review'].str.lower()
```

```python
import re
df['review'] = df['review'].apply(lambda x: re.sub(r'<.*?>', '', x))
```

```python
import string

df['review'] = df['review'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))

#string.punctuation: noktalama işaretleri
```

```python
df['review'].iloc[0]
```

```python
#tokenization

!pip install nltk
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

word_tokenize(df['review'].iloc[0])
```

```python
#stopwords

nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

tokens = word_tokenize(df['review'].iloc[0])
filtered_words = [word for word in tokens if word not in stop_words]
print(filtered_words[:20])
```
